```python
from TripAdvisor.Data import sql_credentials
```

```python
#Load Reviews from Azure SQL Database
import pyodbc
server = sql_credentials.server
database = sql_credentials.database
username = sql_credentials.username
password = sql_credentials.password
driver= '{ODBC Driver 17 for SQL Server}'

with pyodbc.connect('DRIVER='+driver+';SERVER='+server+';PORT=1433;DATABASE='+database+';UID='+username+';PWD='+ password) as conn:
    reviews = pd.read_sql("SELECT * FROM dbo.Reviews", conn, index_col='ID')
```

#### It seems we have 46 national parks for our analysis, but notice that one of them is called "Dry Tortugas National Park Day Trip by Catamaran from Key West", while we are already having "Dry Tortugas National Park" in the list. So we would delete those review entries related to the day trip since those are not related to the park per se. And that should leave us with 45 national parks to analyze.

```python
reviews.drop(reviews[reviews['NationalParkName'] == 'Dry Tortugas National Park Day Trip by Catamaran from Key West'].index, inplace = True)
```

#### Now let's visualize the top 5 and last 5 national parks based on number of reviews

```python
park_counts = reviews['NationalParkName'].value_counts()
```

```python
top5 = park_counts.head().reset_index()
top5.columns = ['ParkName', 'Counts']
```

```python
fig = px.bar(top5, 'ParkName', 'Counts', color='Counts', 
             color_continuous_scale='Redor', title='Top 5 National Parks by Number of Reviews')
fig.show()
```

It seems Dry Tortugas Nation Park has the most reviews. Hmm, that's interesting because I did not expect that. Why? Because normally, without knowing any addtional information, I would assume that the more visits a national parks have, the more reviews there will be.
